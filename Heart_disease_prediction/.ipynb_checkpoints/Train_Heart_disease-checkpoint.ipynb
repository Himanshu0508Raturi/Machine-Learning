### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Dataset

```python
dataset = pd.read_csv('Heart_Failure_Prediction_Dataset.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
dataset
```

```python
x
```

```python
age_column = x[:, 0]
gender_column = x[:,1]
chestP_column = x[:,2]
bp_column = x[:,3]
Cholesterol_column = x[:,4]
bs_column = x[:,5]
ecg_column = x[:,6]
heart_rate_column = x[:,7]
angina_column = x[:,8]
op_column = x[:,9]
slop_column = x[:,10]
```

### Data Preprocessing

### Label Encoding

```python
# Define your DataFrame (assuming it's not defined before)
df = pd.DataFrame({
    'age': age_column,
    'gender': gender_column,
    'chestPain': chestP_column,
    'bp': bp_column,
    'Cholesterol': Cholesterol_column,
    'bsugar': bs_column,
    'ecg': ecg_column,
    'heart_rate': heart_rate_column,
    'angina': angina_column,
    'oldPeak': op_column,
    'SLOPE':slop_column
})
# Save the DataFrame as a JSON file
df.to_json('training_data.json', orient='records', lines=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
```

```python
for row in x:
    print(row)
```

```python
y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x,y,test_size = 0.2 , random_state = 1)
```

```python
len(X_train) , len(X_test) , len(Y_train) , len(Y_test) 
```

```python
(X_train) , (X_test) , (Y_train) , (Y_test) 
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

### Training Dataset

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_naive_bayes = naive_bayes.predict(X_test)
naive_bayes_acc = accuracy_score(Y_test , y_pred_naive_bayes)
print("Accuracy Of Naive Bayes Model: ",naive_bayes_acc)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_randomforest = RandomForestClassifier()
model_randomforest.fit(X_train,Y_train)
```

```python
Y_pred_randomforest = model_randomforest.predict(X_test)
random_forest_acc = accuracy_score(Y_test,Y_pred_randomforest)
print("Accuracy Of Random Forest Model: ",random_forest_acc)
```

### Saving Model

```python
import joblib
joblib.dump(naive_bayes, 'naive_bayes_model.joblib')
joblib.dump(model_randomforest , 'random_forest_model.joblib')
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix , classification_report
cm = confusion_matrix(Y_test , y_pred_naive_bayes)
cm
```

### Classification Report

```python
report = classification_report(Y_test , y_pred_naive_bayes)
print(report)
```
