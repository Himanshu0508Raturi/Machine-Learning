### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Dataset

```python
dataset = pd.read_csv('heart.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Assuming x is a numpy array
x_df = pd.DataFrame(x, columns=dataset.columns[:-1])  # Use dataset's column names excluding the target column
#print(x_df)
print(x_df.head())
```

### Data Preprocessing

### Label Encoding

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x,y,test_size = 0.2 , random_state = 0)
```

```python
len(X_train) , len(X_test) , len(Y_train) , len(Y_test) 
```

```python
X_train
```

```python
xtrain_df = pd.DataFrame(X_train, columns=dataset.columns[:-1])  # Use dataset's column names excluding the target column
print(xtrain_df)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
xtrain_label_df = pd.DataFrame(X_train, columns=dataset.columns[:-1])  # Use dataset's column names excluding the target column
print(xtrain_label_df)
```

```python
X_test
```

### Training Dataset

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_naive_bayes = naive_bayes.predict(X_test)
naive_bayes_acc = accuracy_score(Y_test , y_pred_naive_bayes)
print("Accuracy Of Naive Bayes Model: ",naive_bayes_acc)
```

### Saving Model

```python
import joblib
joblib.dump(naive_bayes, 'naive_bayes_model')
```

### Confusion Matrix(NB)

```python
from sklearn.metrics import confusion_matrix , classification_report , ConfusionMatrixDisplay
cm = confusion_matrix(Y_test , y_pred_naive_bayes)
print(cm)
```

### Displaying Confusion Matrix

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])  # Use display_labels for custom class names
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

### Classification report

```python
report = classification_report(Y_test , y_pred_naive_bayes)
print(report)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(
    min_samples_leaf=5,
    random_state=42,
    bootstrap=True
)
random_forest.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_rf = random_forest.predict(X_test)
rf_acc = accuracy_score(Y_test , y_pred_rf)
print("Accuracy Of Random Forest Model: ",rf_acc)
```

### Confusion matrix and classification report

```python
from sklearn.metrics import confusion_matrix , classification_report , ConfusionMatrixDisplay
cm1 = confusion_matrix(Y_test , y_pred_rf)
print(cm1)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=[0, 1])  # Use display_labels for custom class names
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
report = classification_report(Y_test , y_pred_rf)
print(report)
```

### Saving Model

```python
import joblib
joblib.dump(random_forest, 'random_forest_model')
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier , export_text
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train,Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

```python
from sklearn.metrics import accuracy_score
y_pred_dt = dt.predict(X_test)
dt_acc = accuracy_score(Y_test , y_pred_dt)
print("Accuracy Of Decision Tree Model: ",dt_acc)
```
